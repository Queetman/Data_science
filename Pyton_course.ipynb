6) Показать распределение сотрудников по зарплатам в каждом
департаменте по отдельности

```python
dep_name = df.groupby('department').mean()
dep_name['satisfaction_level'] #Получение названий департаментов
```

```python
dep_name_list = list(dep_name.index)# перевод названий в list
dep_name_list
```

```python
# Написание процедуры для полчения датафрейма с распределением ровня зарплаты в конкретном департаменте.
def get_department_distribution (df,name):
  df_out = df[df['department'] ==name].groupby('salary').count()
  return df_out ['satisfaction_level']
```

```python
dep_distr = pd.DataFrame()
```

```python
# получение датафрейма с
for department in dep_name_list:
  df_dep = get_department_distribution(df,department)
  df_dep_lst = list(df_dep)
  dep_distr.insert (loc= 0 , column=department, value=df_dep_lst)
```

```python
dep_distr
```

```python
dep_distr['salary'] = ['high','low','medium'] # добавление столбца  уровнем з/п
dep_distr = dep_distr[['salary'] + [x for x in dep_distr.columns if x != 'salary']] # перемещения столбца 'saslary'  в начало

dep_distr
```

```python
# Проведение Т-теста (t-критерий тьюдента) Проверка того,что сотрудники с большим окладом проводят больше времень альфа= 5%
result =ttest_ind(df_high['average_montly_hours'],df_low['average_montly_hours'])
print ('p-value: %8f' % result.pvalue)

if ( result.pvalue<alpha):
  print('Отвергаем нулевую гипотезу')
else:
  print('Принимаем нулевую гипотезы')
```

9) Разделить данные на тестовую и обучающую выборки
Построить модель LDA, предсказывающую уволился ли
сотрудник на основе имеющихся факторов (кроме department и
salary)
Оценить качество модели на тестовой выборки

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df.head(1)
```

```python
X = df[['satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company','Work_accident','promotion_last_5years']]
Y = df['left']
```

Разделяем данные на выборку для обучения/проверки

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.1)
```

Создаем и обучаем модель линейной регресии

```python
# Создайте и обучите модель
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.coef_
```

```python
# Визуальная оценка что получилось и что предсказывает модель
out = pd.DataFrame([y_test.values, lr.predict(X_test)]).T
out
# Модель обучилась.
# Есть один момент - параметр уволился сотрудник или нет - это бинарный параметр т.е. он дает либо значение 0, либо 1
# При визуальной оценке видно , что при единиице предсказываемое значение больше, чем при 0.
```

```python
out_working = out [out[0]==1]
out_working[1].mean()
```

```python
out_working[1].std()
```

```python
out_fired = out [out[0]==0]
out_fired[1].mean()
```

```python
out_fired[1].std()
```

```python
# Видно, что среднее значение модели уволившихся и не увольвшихся сотрудников отличается почти в 2 раза 0.38 и 0.18
#СКО приблизительно одинаковые - 0.17-0.19. Если посчитать доверительный интервал, то они у уволившихся
#и не уволившихся сотрудников они будут пересекаться. Т.е. модель не работает со сто процентной вероятностью.
```

```python
# Получите предсказание
lr.predict(X_test)
```

```python
# Оцениваем качество, при помощи метода mean_squared_error для тестовой выборки
mean_squared_error(y_test, lr.predict(X_test))
```
